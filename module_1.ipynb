```python
#calculating student grades
math=float(input("enter ur score"))
soc=float(input("enter ur score"))
eng=float(input("enter ur score"))
sci=float(input("enter ur score"))
hin=float(input("enter ur score"))
avg=float((math+soc+eng+sci+hin)/5)
if avg>90 :
  print("congrats ur grade is A!")
elif avg>80 and avg<90 :
  print("congrats ur grade is B!")
elif  avg>70 and avg<80 :
  print("congrats ur grade is C!")
elif avg>60 and avg<70 :
  print("congrats ur grade is D!")
elif avg<60 :
  print("fail")
```

```python
#Tuples
a=("HI","hello","cherry","HI")
y=a.count("")
print(y)
print(a)
print(len(a))
print(type(a))
```

```python
#various datatypes in tuples
b=("hello","thanks","think")
print(b)
print(len(b))
print(type(b))

c=(1,2,3,4)
print(c)
print(len(c))
print(type(c))

d=(True,False)
print(d)
print(len(d))
print(type(d))

e=('a','b','c')
print(e)
print(len(e))
print(type(e))

f=(1.2,2.3,3.5)
print(f)
print(len(f))
print(type(f))
g=('a',1,2.4,"Tuples",True)
print(g)
print(len(g))
print(type(g))
```

```python
#accessing Tuple items
#negative indexing
g=('a',1,2.4,"Tuples",True,"add")
print(g[4])
print(g[-4])
print(g[1:4])
print(g[0:4])
print(g[:3])
print(g[1:])
```

```python
#checking items in the the tuple
g=('a',1,2.4,"Tuples",True,"add")
if 2.4 in g:
  print("yes")
else :
  print("No")
```

```python
#inserting the items in a tuples
g=('a',1,2.4,"Tuples",True,"add")
h=list(g)
h[3]=3.544444
g=tuple(h)
print(h)
```

```python
#using append
g=('a',1,2.4,"Tuples",True,"add")
h=list(g)
h.append(14563413216541546541653146546546768358417684867)
g=tuple(h)
print(h)
```

```python
#loops in tuples
g=('a',1,2.4,"Tuples",True,"add")
for i in range(1,4):
  print(i)
```

```python
#sets
g={'a',1,2.4,"Tuples",True,"add"}
print(g)
for i in g:
  print(i)
```
